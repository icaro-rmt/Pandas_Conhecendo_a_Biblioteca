Qual o Percentual de cada tipo de imovel na base de dados?

```python
df.Tipo.value_counts(normalize=True)
```

```python
df.Tipo.value_counts(normalize=True).to_frame()
```

```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_por_tipo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo', ascending=True)
```

```python
df_percentual_por_tipo.plot(kind='bar', figsize=(10,8), color='green', edgecolor='black',xlabel='Tipos',ylabel='Percentual')
```
